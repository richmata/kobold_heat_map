```python
from itertools import combinations

import pandas as pd
import geoplot as gplt
import geopandas as gpd
from shapely.ops import unary_union, polygonize, cascaded_union
from shapely.geometry import LineString, MultiLineString, Polygon, MultiPolygon, LinearRing

"""
# # Steps to success: # #
-------------------------
# classify polygons as containing either serpentinite, granodiorite, or None
# if polygons contain either serpentinite or granodiorite, find areas of intersection
# create a planar graph of all intersecting polygons with optional buffer zone
# plot polygons as a headmap with designated boundary
"""

class BedrockHeatMap:
    CLASSIFICATIONS = ["S", "G"]
    
    def __init__(self, buffer: float=5000.0, file_name: str="data/BedrockP.shp"):
        self.buffer = buffer
        self.data =  self._data(file_name) # gpd.read_file(file_name)
        self.boundary = gpd.GeoSeries(LineString(unary_union(self.data["geometry"]).exterior))
      
    # TODO: remove
    def _data(self, file: str) -> gpd.GeoDataFrame: 
        df = gpd.read_file(file)
        df["class"] = df.apply(self._classify, axis=1)
        return df

    def _classify(self, row) -> str or None: # TODO: find a better way to classify using strat unit/name/age
        """
        Returns classification of bedrock for a given location
        """
        serpentinite_keywords = ["serpentinite", "ultramafic"] # add extra conditions
#         serpentinite_age = ["Mississippian", "Late Devonian"]
        granodiorite_keywords = ["granodiorite", "volcanic"] # add extra conditions
    
        if row["strat_name"] is None:
            return
        if any(keyword in row["strat_name"].lower() for keyword in serpentinite_keywords):
            return self.CLASSIFICATIONS[0]
        if (
            any(keyword in row["strat_name"].lower() for keyword in granodiorite_keywords)
            or row["strat_age"].lower() == "lower cretaceous"
        ):
            return self.CLASSIFICATIONS[1]
        return
    
    def _probability(self, row) -> float:
        """
        Returns probabilty of finding cobalt deposit at given location
        """
        cobalt_area = sum(cobalt.area for cobalt in row["cobalt_deposits"])
        return cobalt_area / row["area_m2"]
        
        # apply Bayes Theorem:
        # # P(C|S+G) = P(C)P(S+G|C)/P(P(S+G)
        
#         total_land_area = pd.Series(self.data["area_m2"]).sum()
#         total_cobalt_area = pd.Series(self.planar_graph["area"]).sum()
#         total_evidence_area = pd.Series(self.possible_cobalt_deposits["area_m2"]).sum()
        
#         P_c = total_cobalt_area / total_land_area # P(C): probabilty of cobalt
#         P_sg = total_evidence_area / total_land_area # P(S+G): probability of serpentinite and granodiorite

        # find sum of all areas of cobalt within the given boundary
#         area_sum = 0
#         for polygon in self.cobalt_deposits["geometry"]:
#             if row["geometry"].intersects(polygon):
#                 area_sum += polygon.area
        
#         probability = area_sum / row["area_m2"] # P(C|S+G): probablity of finding cobalt in given location
#         return probability
#         return (P_c * P_sg_c) / P_sg # P(C|S+G): probability of finding cobalt in a given location
        
    @property
    def cobalt(self) -> gpd.GeoDataFrame:
        """
        Returns dataframe with probabilities of finding cobalt at a given location
        """
        df = self.data
#         df["class"] = df.apply(self._classify, axis=1)
        df["cobalt_deposits"] = df.apply(self._cobalt_deposits, axis=1)
        df["probability"] = df.apply(self._probability, axis=1)
        return df
        
    @property
    def possible_cobalt_deposits(self) -> gpd.GeoSeries:
        """
        Returns a planar graph of intersecting polygons of possible cobalt deposits
        """
        
        # get areas where serpentinite or granodiorite are located
        df = self.data
        df = df[df["class"].isin(self.CLASSIFICATIONS)]
        
        # find intersections of serpentinite and granodiorite
        intersections = [
            polygon_1.intersection(polygon_2)
            for polygon_1, polygon_2 in combinations(df["geometry"], 2)
            if polygon_1.intersects(polygon_2)
        ]
        
        # cut lines at each intersection
        rings = [] # collection of LineStrings
        for intersection in intersections:
            
            # get list of LineStrings from Polygons
            if isinstance(intersection, Polygon):
                rings.extend([LineString(list(intersection.exterior.coords))])
    
            # get list of LineStrings from MultiLineString
            elif isinstance(intersection, MultiLineString):
                rings.extend([linestring for linestring in intersection])
        
        # create union of rings and polygonize
        union = unary_union(rings)
        deposits = gpd.GeoSeries(
            [
                Polygon(geom.buffer(self.buffer).exterior) # .set_crs(epsg=26910)
                for geom in polygonize(union)
            ]
            # .set_crs(epsg=26910)
        )
        return deposits # gpd.GeoSeries([Polygon(geom.buffer(self.buffer).exterior) for geom in polygonize(union)])
    
#     @property
    def _cobalt_deposits(self, row) -> gpd.GeoSeries:
        """
        Returns all cobalt deposits with a given location
        """
        # get areas where serpentinite or granodiorite are located
#         df = self.possible_cobalt_deposits.set_crs(epsg=4326, allow_override=True)
        
#         # find intersections of serpentinite and granodiorite
#         intersections = [
#             polygon_1.intersection(polygon_2)
#             for polygon_1, polygon_2 in combinations(self.possible_cobalt_deposits["geometry"], 2)
#             if polygon_1.intersects(polygon_2)
#         ]
        
#         # cut lines at each intersection
#         rings = [] # collection of LineStrings
#         for intersection in intersections:
            
#             # get list of LineStrings from Polygons
#             if isinstance(intersection, Polygon):
#                 rings.extend([LineString(list(intersection.exterior.coords))])
    
#             # get list of LineStrings from MultiLineString
#             elif isinstance(intersection, MultiLineString):
#                 rings.extend([linestring for linestring in intersection])
        
#         # create union of rings and polygonize
#         union = unary_union(rings)
#         deposits = [
#             Polygon(geom.buffer(self.buffer).exterior) # .set_crs(epsg=26910)
#             for geom in polygonize(union)
#         ] # set as geoseries
        
        cobalt_deposits = []
        for deposit in self.possible_cobalt_deposits:
            if row["geometry"].intersects(deposit):
                cobalt_deposits.append(deposit)
        return cobalt_deposits
        
#         return gpd.GeoDataFrame({"geometry": gpd.GeoSeries(planar), "area": planar.area}, crs="EPSG:26910")

import time
start_time = time.time()
BHM = BedrockHeatMap() # want to take buffer value as parameter when running with click
cobalt = BHM.cobalt
print("--- %s seconds ---" % (time.time() - start_time))


# plot head map
# BHM.planar_graph.append(BHM.boundary).plot(color = "red", figsize=(20,20))
# df.plot(column="probability", categorical=True, legend=True, figsize=(20,20), cmap="RdYlGn")
```
