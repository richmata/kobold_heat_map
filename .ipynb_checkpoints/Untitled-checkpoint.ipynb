```python
import geopandas as gpd

def _classification(row: str) -> str or None:
    if row["strat_name"] is None:
        return
    if "serpentinite" in row["strat_name"].lower() or "ultramafic" in row["strat_name"].lower():
        return "serp"
    elif "volcanic" in row["strat_name"].lower() or row["strat_age"].lower() == "lower cretaceous":
        return "gran"
    return
    

gdf = gpd.read_file("data/BedrockP.shp")
gdf["classification"] = gdf.apply(_classification, axis=1)
gdf[gdf["classification"] == "gran"].count()
```
