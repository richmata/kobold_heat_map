```python
from itertools import combinations

import geoplot as gplt
import geopandas as gpd
from shapely.ops import unary_union, polygonize, cascaded_union
from shapely.geometry import LineString, MultiLineString, Polygon, MultiPolygon

"""
# # Steps to success: # #
-------------------------
# classify polygons as containing either serpentinite, granodiorite, or None
# if polygons contain either serpentinite or granodiorite, find areas of intersection
# create a planar graph of all intersecting polygons with optional buffer zone
# plot polygons as a headmap with designated boundary
"""

class BedrockHeatMap:
    CLASSIFICATIONS = ["S", "G"]
    
    def __init__(self, buffer: float=5000.0, file_name: str="data/BedrockP.shp"):
        self.buffer = buffer
        self.data = self._data(file_name)
        self.boundary = gpd.GeoSeries(LineString(cascaded_union(self.data["geometry"]).exterior))
        
    @classmethod
    def _data(cls, file: str) -> gpd.GeoDataFrame:
        df = gpd.read_file(file)
        df["class"] = df.apply(cls._classify, axis=1)
        return df

    @classmethod
    def _classify(cls, row) -> str or None: # TODO: find a better way to classify using strat unit/name/age
        serpentinite_keywords = ["serpentinite", "ultramafic"]
        granodiorite_keywords = ["granodiorite", "volcanic"]
    
        if row["strat_name"] is None:
            return
        if any(keyword in row["strat_name"].lower() for keyword in serpentinite_keywords):
            return cls.CLASSIFICATIONS[0]
        if (
            any(keyword in row["strat_name"].lower() for keyword in granodiorite_keywords)
            or row["strat_age"].lower() == "lower cretaceous"
        ):
            return cls.CLASSIFICATIONS[1]
        return
    
    @staticmethod
    def intersections(geoms: gpd.GeoSeries) -> list:
        """
        Returns a list of intersecting polygon combinations
        """
        intersections = [
            polygon_1.intersection(polygon_2)
            for polygon_1, polygon_2 in combinations(geoms, 2)
            if polygon_1.intersects(polygon_2)
        ]
        return intersections
    
    @property
    def cobalt(self) -> gpd.GeoSeries:
        """
        Returns potential locations of cobalt deposits
        """
        
        # get areas where serpentinite and granodiorite are located
        df = self.data
        df = df[df["class"].isin(BHM.CLASSIFICATIONS)]
        
        # find intersections of serpentinite and granodiorite
        intersections = self.intersections(df["geometry"])
        
        # cut lines at each intersection to create a planar graph of intersecting Polygons
        rings = [] # collection of LineStrings
        for intersection in intersections:
            
            # get list of LineStrings from Polygons
            if isinstance(intersection, Polygon):
                rings.extend([LineString(list(intersection.exterior.coords))])
    
            # get list of LineStrings from MultiLineString
            elif isinstance(intersection, MultiLineString):
                rings.extend([linestring for linestring in intersection])
        
        # create union of polygons as a multipolygon
        union = unary_union(rings)
        planar = [Polygon(geom.buffer(self.buffer).exterior) for geom in polygonize(union)]        
        return gpd.GeoSeries(MultiPolygon(planar))


BHM = BedrockHeatMap() # want to take buffer value as parameter when running with click

# combine potential locations of cobalt with bedrock bounary 
cobalt = BHM.cobalt
graph = cobalt.append(BHM.boundary, ignore_index=True)

# need to calculate a value to use for heatmap (0-1, 0 means no cobalt, 1 cobalt exists)
# try taking center point of polygons in multipolygon
# check if points in original dataset are found within intersecting polygons within a threshold

# plot head map
graph.plot(color = "red", figsize=(20,20))
# graph.plot(column="strat_unit", categorical=True, legend=True, figsize=(20,20), cmap="RdYlGn")
```
